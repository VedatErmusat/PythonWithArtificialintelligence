```python
import pandas as pd
```

```python
df = pd.read_csv("OneDrive/Masaüstü/vedat/CODİNG/AI_PYTHON/PythonVeriAnalizProjelerim/imdb_top_1000.csv")
```

```python
df.head(3)
```

```python
# Poster linki sütununu kaldırma
df = df.drop('Poster_Link', axis=1) # Satır içerisinden bir şey atmak istendiği zaman axis=0 derdik
```

```python
df.head()
```

```python
df["Runtime"]
```

```python
# Min yazılarını kaldırmamız gerek
df["Runtime"].str.replace(" min","")
```

```python
df.info()
```

```python
# Runtime verileri hala string, sayısal değere çevirmemiz gerek
df["Runtime"] = df["Runtime"].str.replace(" min","").astype("int64")
```

```python
df.info()
```

```python
# Gross değerlerimiz de hala string gözüküyor, boş olan yani Nan değerleri kaldırıp sayısal değere çevirelim
df = df.dropna(subset=["Gross"])
```

```python
df["Gross"] = df["Gross"].str.replace(",","").astype("float64")
```

```python
# Dakikalık kazancını hesaplayalım
df["Dolar/Min"] = df["Gross"]/df["Runtime"]
```

```python
df.head(4)
```

```python
df.info()
```

```python
# Filmleri dakikalık kazançlarına göre büyükten küçüğe sıralayalım
df.sort_values(by="Dolar/Min",ascending = False)
```
